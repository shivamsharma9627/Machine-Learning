```python
import pandas as pd
df=pd.read_csv("D:\\Documents\Bengaluru_House_Data.csv")
df.dropna(inplace=True)
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, "price"] > 150:
        df.loc[x, "price"] = 1
    else:
        df.loc[x, "price"] = 0
```

```python
df
```

```python
inputs=df.drop('price',axis='columns')
target=df['price']
```

```python
from sklearn.preprocessing import LabelEncoder
le_area_type = LabelEncoder()
le_availability = LabelEncoder()
le_location = LabelEncoder()
le_size= LabelEncoder()
le_society = LabelEncoder()
le_total_sqft = LabelEncoder()
le_bath= LabelEncoder()
le_balcony = LabelEncoder()
```

```python

inputs['area_type_n']=le_area_type.fit_transform(inputs['area_type'])
inputs['availability_n']=le_availability.fit_transform(inputs['availability'])
inputs['location_n']=le_location.fit_transform(inputs['location'])
inputs['size_n']=le_size.fit_transform(inputs['size'])
inputs['society_n']=le_society.fit_transform(inputs['society'])
inputs['total_sqft_n']=le_total_sqft.fit_transform(inputs['total_sqft'])
inputs['bath_n']=le_bath.fit_transform(inputs['bath'])
inputs['balcony_n']=le_balcony.fit_transform(inputs['balcony'])
inputs.head()
```

```python
inputs_n=inputs.drop(['area_type','availability','location','size','society','total_sqft','bath','balcony'],axis='columns')
```

```python
inputs_n
```

```python
target[40:60]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n,target,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
